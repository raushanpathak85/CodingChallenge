```python
# /**
#  * Have the function PrimeMover(num) return the numth prime number. The range will be from 1 to 10^4.
#  * For example: if num is 16 the output should be 53 as 53 is the 16th prime number.
#  *
#  * 
#  */

def prime_mover(num):
    
    if num==1:
        return 2
    if num==2:
        return 3
    primes=1
    
    for i in range(3, 100000):
        prime=0
        for j in range (2,i):
            if i%j == 0:
                prime=prime+1
                
        if prime ==0:
            primes=primes+1
        
        if primes == num:
            return i
        
```

```python
prime_mover(16)
```
