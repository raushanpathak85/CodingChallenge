```python
def is_balanced(param):
    stack=[]
    opening=set({'(','[','}'})
    closing=set({')',']','}'})
    #dict1={}
    pairs=dict({')':'(', ']':'[','}':'{'})
    
    for c in param:
        if c in opening:
            stack.append(c)
        elif c in closing:
            if not stack or stack[-1] != pairs[c]:
                return False
        #stack.pop()
    return len(stack)==0
```
